```python
##scatter chart(산점도)
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../sample.csv')

plt.figure(num="성적 그래프")

# plt.rcParams.update({'font.family':'Malgun Gothic', 'font.size':12})
plt.rcParams.update({'font.family' : 'AppleGothic', 'font.size':14})
plt.title("학번별 성적")


ax=plt.gca()
xcolumns = ['학번', '학번', '학번']
ycolumns = ['국어', '영어', '수학']

color = ['#2091ff', '#99aacd', '#ff00ff', '#00ffff', '#0000ff', '#ffffff'] #산점도의 점 색상
data.plot(kind='scatter', x='학번', y='국어', ax=ax, color=color[1], s=50)
data.plot(kind='scatter', x='학번', y='영어', ax=ax, color=color[2], s=50)
data.plot(kind='scatter', x='학번', y='수학', ax=ax, color=color[4], s=100)
#s => 점에 대한 사이즈 color => 점에 대한 색상

plt.xlabel("학번")
plt.ylabel("성적")

plt.show()
```

```python
##pychart 만들기 pychart
import pandas as pd
import matplotlib.pyplot as plt

# data = pd.read_csv('../sample.csv')
data = pd.DataFrame({'mass':[0.330, 4.87, 5.97], 'radius':[2439.7, 6051.8, 4483.2]})
index=['Mercury', 'Venus', 'Earth']

plt.figure(num="py Chart")

# plt.rcParams.update({'font.family':'Malgun Gothic', 'font.size':12})
plt.rcParams.update({'font.family' : 'AppleGothic', 'font.size':14})
plt.title("행성 크기 비교")

ax=plt.gca()

data.plot(kind='pie', y='radius', ax=ax)
#pie => 그래프의 종류

plt.xlabel("학번")
plt.ylabel("radius")

plt.show()
```

```python
#histogradm만들기 (도수분포도)
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../sample.csv')
data = data.drop('학번', axis=1)

plt.figure(num="히스토그램")

plt.rcParams.update({'font.family' : 'AppleGothic', 'font.size':14})
plt.title("과목별 점수 빈도")

ax = plt.gca()

data.plot(kind='hist',by=None, bins=20, ax=ax, alpha=0.8, edgecolor='blue', linewidth=2)
#by의 쓰임 alpha=> 투명도 bins=?
plt.xlabel("과목별 성적")
plt.ylabel("빈도수")
plt.show()
```
