```python
###*** 과제내용 중요
def Func(row):
    if row > 90:
        return 'A'
    elif row >80:
        return 'B'
    else:
        return 'C'

data = pd.read_csv("sample.csv")
data['총합'] = data['수학']+data['국어']+data['영어']
data['평균'] = data['총합']/3
data['등수'] = data['총합'].rank(ascending=False)
data = data.set_index('등수')
data = data.sort_index()
data['학점'] = data['평균'].apply(Func)

data.sort_values(axis=0, by='총합', ascending=1)
print(data)
```
